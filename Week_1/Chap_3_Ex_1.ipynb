```python
import pyspark
from pyspark import SparkContext
sc = SparkContext(master="local", appName="New Spark Context")
```

```python
# 1
rdd = sc.textFile('Resources_day_2/kddcup.data_10_percent.gz')
```

```python
# 2
print('2', rdd.count())
```

```python
# 3
print('3', rdd.take(3))
```

```python
# 4
rdd_normal = rdd.filter(lambda x: 'normal' in x)
```

```python
# 5
print('5', rdd_normal.count())
```

```python
# 6
array = rdd.collect()
```

```python
# 7 
rdd_sample = rdd.sample(False, 0.1, seed=42)
```

```python
# 8
rdd_normal_sample = rdd_normal.intersection(rdd_sample)
print('8', rdd_normal_sample.count()/ rdd_sample.count())
```

```python
# 9
rdd_sample_not_normal = rdd_sample.subtract(rdd_normal_sample)
print('9', rdd_sample_not_normal.count())
```

```python
# 10
array_RDD = rdd.map(lambda s: s.split(','))
```

```python
# 11
protocols = array_RDD.map(lambda x: x[1]).distinct()
protocols.collect()
```

```python
# 12
service = array_RDD.map(lambda x: x[2]).distinct()
service.collect()
```

```python
# 13
product = protocols.cartesian(service).collect()
product
```
