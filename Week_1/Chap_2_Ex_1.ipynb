```python
import pyspark
from pyspark import SparkContext
sc = SparkContext(master="local", appName="New Spark Context")
```

```python
data = sc.textFile('Resources_day_1/t8.shakespeare.txt')
data.count()
```

```python
data.filter(lambda s: 'love' in s).count()
```

```python
import re
data.filter(lambda s: re.search('\d+', s)).count()
```

```python
import re
data.filter(lambda s: re.match('.*\d+.*', s)).count()
```

```python
contain_number = data.filter(lambda s: re.match('.*\d+.*', s))
print('Total number of line: ', contain_number.count())
i = 1
for line in contain_number.collect():
    print(i, ': ', line)
    i += 1
```
